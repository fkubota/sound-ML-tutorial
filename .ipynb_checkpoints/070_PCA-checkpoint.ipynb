# Let's import everything we nead :)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

# Data Preparation

```python
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target
a = np.arange(0, len(y))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.33, 
                                                    random_state=42)
```

# Exploratory Data Analysis

**plot**
- 特徴量ごとにプロットしてみます

```python
feat_names = cancer.feature_names

fig, axs = plt.subplots(10, 3, figsize=(20, 20))
axs = axs.ravel()
for i in range(len(feat_names)):
    feat0 = X[:, i][y==0]
    a0 = a[y==0]
    feat1 = X[:, i][y==1]
    a1 = a[y==1]
    
    axs[i].scatter(a0, feat0, s=1, c='blue')
    axs[i].scatter(a0, feat0, s=1, c='blue')
    
```
