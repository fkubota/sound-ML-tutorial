# Let's import everything we nead :)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

# Data Preparation

```python
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target
a = np.arange(0, len(y))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.33, 
                                                    random_state=42)
```

# Exploratory Data Analysis

**plot**
- 特徴量ごとにプロットしてみます

```python
feat_names = cancer.feature_names

fig, axs = plt.subplots(10, 3, figsize=(20, 20))
axs = axs.ravel()
for i in range(len(feat_names)):
    feat0 = X[:, i][y==0]
    a0 = a[y==0]
    feat1 = X[:, i][y==1]
    a1 = a[y==1]
    
    axs[i].scatter(a0, feat0, s=1, c='blue')
    axs[i].scatter(a1, feat1, s=1, c='orange')
    axs[i].set_ylabel(feat_names[i])
    
```

---> 横軸がデータのインデックス。縦軸が、その特徴量の大きさを示しています。  
---> 色が、上下で別れている特徴量ほど、分離する際に参考になる良い特徴量ということになります。  
---> 左下の `worst concave points` なんかはかなりはっきりと別れていますね。その上の`worst smoothness`は明らかに分離がうまくいっていません。
---> このデータは、30 もの特徴量が有りますが、全て使わなくてもそこそこ良い結果が出そうです。  
---> PCA を使って次元削減をしましょう。  
---> PCA は、単に特徴量を減らすような手法ではないことに注意してください。
